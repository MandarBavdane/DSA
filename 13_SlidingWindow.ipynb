# 1. Variable Length Sliding Window

```python
# Given a string s, find the length of longest substring without repeating characters
s = "abcabcbb"

def lenghOfLongestString(s):
    l = 0
    longest = 0
    sett = set()
    n = len(s)
    
    # O(n)
    for r in range(n): 
        while s[r] in sett: # run n times total, not per iteration r
            sett.remove(s[l])
            l += 1
        
        w = (r - l) + 1
        longest = max(longest, w)
        sett.add(s[r])
        
    return longest
    # Time: O(n)
    # Space: O(n)

print(lenghOfLongestString(s))        
```

# Fixed Lenght Sliding Window

```python
k = 4
nums = [1, 12, -5, -6, 50, 3]
def findMaxAverage(nums, k):
    n = len(nums)
    cur_sum = 0
    
    for i in range(k):
        cur_sum += nums[i]
        
    max_avg = cur_sum / k
    
    for i in range(k, n):
        cur_sum += nums[i]
        cur_sum -= nums[i-k]
        
        avg = cur_sum/k
        max_avg = max(max_avg, avg)
        
    return max_avg

    # Time: O(n)
    # Space: O(1)

print(findMaxAverage(nums, k))
```

```python
# Build Min Heap (Heapify)
# Time: O(n), Space: O(1)

A = [-4, 3, 1, 0, 2, 5, 10, 8, 12, 9]

import heapq
heapq.heapify(A) # gives min-heap (sorts in ascending)

A
```

```python
# Heap Push (Insert element)
# Time: O(log n)

heapq.heappush(A, 4)

A
```

```python
# Heap Pop (Extract Min)
# Time: O(log n)

minn = heapq.heappop(A)

A, minn
```

```python
# Heap Sort
# Time: O(n log n) # One of the best complexity for sorting algo, Space: O(1)/O(n)
# note: O(1) Space is possible via swapping, but this is complex

def heapsort(arr):
    heapq.heapify(arr)
    n = len(arr)
    new_list= [0] * n
    
    for i in range(n):
        minn = heapq.heappop(arr)
        new_list[i] = minn
        
    return new_list

heapsort([1,3, 5, 7, 9, 2, 4, 6, 8, 0])
```

```python
# Heap Push Pop: Time: O(log n)
heapq.heappushpop(A, 99)
A
```

```python
# Peek at Min: Time: O(1)
A[0]
```

```python
# Max Heap

B = [-4, 3, 1, 0, 2, 5, 10, 8, 12, 9]
n = len(B)

for i in range(n):
    B[i] = -B[i]
    
heapq.heapify(B)

B
```

```python
largest = -heapq.heappop(B)

largest
```

```python
heapq.heappush(B, -7) # Insert 7 into max heap

B
```

```python
# Bilud heap from scratch

C = [-5, 4, 2, 1, 7, 0 ,3]

heap = []

for x in C:
    heapq.heappush(heap, x)
    print(heap, len(heap))
```

```python
# Putting tuples of items on the loop

D = [5, 4, 3, 5, 4, 3, 5, 5, 4]

from collections import Counter

counter = Counter(D) # to get the frequencies of elements in the array
                     # key(element) : value(frequency)

counter
```

```python
for k, v in counter.items():
    print(k, v)
```

```python
heap = []

for k, v in counter.items():
    heapq.heappush(heap, (v, k)) # value(frequency) : key(element)
    
heap
```

```python
heap = []

for k, v in counter.items():
    heapq.heappush(heap, (v, k)) # value(frequency) : key(element)
    
heap
```

```python
sorted(heap)
```

```python
# sorted heap can also be formed using while loop
while heap: 
    print(heapq.heappop(heap))
```

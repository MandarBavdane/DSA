```python
A = [-4, -1, 0, 3, 11]

def sortedSquares(A):
    left = 0 
    right = len(A) - 1
    result = [] # first result will look [16, 1, 0, 9, 121] -> [121, 16, 9, 1, 0] and finally on reverse will get the o/p
    
    while left <= right:
        if abs(A[left]) > abs(A[right]):
            result.append(A[left] ** 2)
            left += 1
        else:
            result.append(A[right] ** 2)
            right -= 1
            
    result.reverse()
    return result

print(sortedSquares(A))
```

```python
# Factory to make a human
'''
This is a class that represents a simplified Human.

It takes their age as an int and their name as a string.
'''
class Human:
    def __init__(self, age, name): # __ Dunder/Double Underscore __ is used to indicate that this is a special method, it is called when we create an object of the class, it is used to initialize the attributes of the object
        self._age = age # _(single underscore) This is internal. Don’t access directly unless necessary.
        self._name = name
        
    def __str__(self):
        return f"A human with name {self._name}, their age is {self._age}." 
    
    def __repr__(self): # repr in collab talks with str so no need to print(h), h will give the same output as str(h)
        return f"A human with name {self._name}, their age is {self._age}." 
    

    def older_younger_than(self, age):
        if self._age > age:
            print("Our age is greater than their age.")
        elif self._age == age:
            print("Our age is equal to their age.")
        else:
            print("Our age is less than their age.")
        
h = Human(age=23, name="Mandar")

h
```

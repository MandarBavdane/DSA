```python
# Hashsets

s = set()
print(s)
```

```python
# Add items into set - O(1)
s.add(1)
s.add(2)
s.add(3)

print(s)
```

```python
# Lookup if item is in set - O(1)
if 1 not in s:
    print("1 is in the set")
else:    
    print("1 is not in the set")
```

```python
# Remove item from set - O(1)
s.remove(3)

print(s)
```

```python
# Set Construction - O(string length) 

string = 'aaaaaaaaaaaaaaabbbbbbbbbbbbbccccccccccccccddddddddddddddeeeeeeeeeeeefffffffffffffff'
sett = set(string) # Set Construction

sett
```

```python
# Hashmaps - Dictionaire

d = {'greg': 1, 'mandar': 2, 'rob': 3}

print(d)
```

```python
# Add Key: val in dictionary: O(1)
d['aish'] = 4

print(d)
```

```python
# Check for presence of key in dictionary - O(1)
if 'greg' in d:
    print(True)
```

```python
# Check the value corresponding to a key in the dictionary - O(1)
print(d['greg'])
```

```python
# Loop over the key:val pairs in the dictionary - O(n)
for key, val in d.items():
    print(f'key {key}: val {val}')
```

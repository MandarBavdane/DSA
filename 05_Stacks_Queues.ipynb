```python
# Stacks --> LIFO (Last In First Out)

stk = []

print(stk)
```

```python
# Append to top of stack - O(1)
stk.append(5)
stk
```

```python
stk.append(4)
stk.append(3)

stk
```

```python
# Pop from top of stack - O(1)
x = stk.pop()
print(x)
print(stk)
```

```python
# Ask what's on the top of the stack - O(1)
print(stk[-1])
print(stk)
```

```python
# Ask if something is in the stack/ or see the stack empty? - O(n)
if stk:
    print(True)
else:
    print(False)
```

```python
# Make sure before popping, the stack should not be empty, otherwise it will throw an error.
# To avoid this, we can check if the stack is empty before popping.
```

```python
# QUEUES --> FIFO (First In First Out). 
from collections import deque

q = deque()

print(q)
```

```python
# Enqueue - Add element to the end of the queue - O(1)
q.append(5)
q.append(6) 
q.append(7)
q
```

```python
# Dequeue (pop left) - remove element from the beginning/left of the queue - O(1)
q.popleft() # as pop removes from the end of the list as in stack
q
```

```python
# Peek from left side - O(1)
q[0]
```

```python
# Peek from right side - O(1)
q[-1]
```

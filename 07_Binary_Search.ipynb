```python
A = [-3, -1, 0, 1, 4, 7]

# Naive O(n) searching

if 3 in A:
    print(True)
```

```python
# Traditional Binary Search - Looking up if number is in array:
# Time: O(log n)
# Space: O(1)

# This method only works if the array is sorted. If the array is not sorted, we can sort it first and then apply binary search
def binary_search(arr, target):

    N = len(arr)
    L = 0
    R = N - 1

    while L <= R:
        # M = (L + R) // 2
        M = L + ((R - L) // 2)

        if arr[M] == target:
            return True
        elif target < arr[M]:
            R = M - 1
        else:
            L = M + 1            
            
    return False

binary_search(A, -3)

```

```python
# Based on over under OR condition based technique
B = [False, False, False, False, True, True, True] # It works only on sorted arrays where all False are followed by True

def binary_search_condition(arr):
    N = len(arr)
    L = 0
    R = N - 1
    
    
    while L < R:
        M = L + ((R - L) // 2)
        
        if B[M]:
            R = M
        else:
            L = M + 1
        
    return L

binary_search_condition(B)
```

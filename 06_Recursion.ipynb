```python
# Fibonacci
# F(0) = 0
# F(1) = 1
# for n > 1: F(n) = F(n-1) + F(n-2) 

# Time complexity: O(2^n); Space complexity: O(n) due to the call stack
def F(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return F(n-1) + F(n-2)
    
F(12)
```

```python
# Singly Linked List

class SinglyNode:
    def __init__(self, val, next=None):
        self.val = val
        self.next = next
        
    def __str__(self):
        return str(self.val)
    
Head = SinglyNode(1)
A = SinglyNode(3)
B = SinglyNode(4)
C = SinglyNode(7)

Head.next = A
A.next = B
B.next = C

print(Head)
```

```python
# Linked Lists

# Time: O(n) n is the number of nodes in the linked list; Space: O(n) due to the call stack
def reverse(node):
    if not node:
        return
    
    reverse(node.next)
    print(node)
    
reverse(Head)
```
